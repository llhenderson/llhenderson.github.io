```python
import numpy as np
import re
import pandas as pd
```

```python
df = pd.read_csv("frankenstein_chunks.csv")
text = []
for portion in df.text:
    text += portion

text2 = str(df.text.tolist())
text2 = re.findall(r'\b\w+\b', text2.lower())
text2 = [word for word in text2 if word !='n']
```

```python
def consenant_vowel_count(text):
    vowels = set()
    vowels.add('a')
    vowels.add('e')
    vowels.add('i')
    vowels.add('o')
    vowels.add('u')
    consenant = 0
    vowel = 0
    cc = 0
    cv = 0
    vv = 0
    vc = 0
    for word in text:
        for i in range(len(word)-1):
            if word[i] in vowels:
                vowel += 1
            else:
                consenant +=1
            if word[i] in vowels:
                if word[i+1] in vowels:
                    vv +=1
                    
                else:
                    vc +=1
            elif word[i] not in vowels:
                if word[i+1] in vowels:
                    cv +=1
                    
                else:
                    cc +=1
    return cc,cv,vv,vc,consenant,vowel
cc,cv,vv,vc,consenant,vowel = consenant_vowel_count(text2)
sum = cc + cv + vv + vc
cc_percentage = cc / sum
cv_percentage = cv / sum
vv_percentage = vv / sum
vc_percentage = vc / sum
#The difference between independant and observed probabilities prove that the v-c configurations are dependant events
```

```python
import random
vv_transition_probability = vv_percentage / ( vowel/ (vowel + consenant))
vc_transition_probability = 1 - vv_transition_probability
cc_transition_probability = cc_percentage / ( consenant/ (vowel + consenant))
cv_transition_probability = 1 - cc_transition_probability

iter = 10000

vowel_count = 0
consenant_count = 0
current_type = 'v'

for _ in range(iter):
    roll = random.random()
    if current_type == 'v':
        if roll < vv_transition_probability:
            current_type = 'v'
            vowel_count +=1 
        else:
            current_type = 'c'
            consenant_count +=1 
    elif current_type == 'c':
        if roll < cc_transition_probability:
            current_type = 'c'
            consenant_count +=1
        else:
            current_type = 'v'
            vowel_count += 1
            
observed_consenant_percentage = consenant / (consenant + vowel)
observed_vowel_percentage = vowel / (consenant + vowel)
predicted_consenant_percentage = consenant_count / (consenant_count + vowel_count)
predicted_vowel_percentage = vowel_count / (consenant_count + vowel_count)

print(f'Predicted vowel percentage: {predicted_vowel_percentage} \nObserved vowel percentage: {observed_vowel_percentage}')
print(f'\nPredicted consenant percentage: {predicted_consenant_percentage} \nObserved consenant percentage: {observed_consenant_percentage}')
```
